–Ø –±—ã —Ç–∞–∫ –æ–ø—Ä–µ–¥–µ–ª–∏–ª —Å–ø–æ—Å–æ–± —Ä–µ—à–µ–Ω–∏—è —Ç–∞–∫–æ–π –∑–∞–¥–∞—á–∏: –µ—Å–ª–∏ –∑–∞–ø–∏—Å–∏ —Å –ø—Ä–æ–ø—É—Å–∫–∞–º–∏ –≤ –∫–ª—é—á–µ–≤—ã—Ö –¥–ª—è –º–æ–µ–≥–æ –∞–Ω–∞–ª–∏–∑–∞ –ø–æ–ª—è—Ö —Å–æ—Å—Ç–∞–≤–ª—è—é—Ç 1-3-5-–¥–æ 10% —Å—Ç—Ä–æ–∫, —Ç–æ —è –±—ã –∏—Ö –∏–∑ –æ—Ä–∏–≥–∏–Ω–∞–ª—å–Ω–æ–≥–æ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ –≤—ã—Ä–µ–∑–∞–ª (–Ω–æ —Å–æ—Ö—Ä–∞–Ω–∏–ª –±—ã –≤ –Ω–µ–∫–∏–π df, –∫ –∫–æ—Ç–æ—Ä–æ–º—É –ø—Ä–∏ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ—Å—Ç–∏ –º–æ–≥ –±—ã –æ–±—Ä–∞—Ç–∏—Ç—å—Å—è).

–ê –≤–æ–æ–±—â–µ, –≤ —Ç–∞–∫–∏—Ö —Å–ª—É—á–∞—è—Ö, –∫–æ–≥–¥–∞ –Ω—É–∂–Ω–æ —Ä–µ—à–∏—Ç—å - –∑–∞–ø–æ–ª–Ω—è—Ç—å –ø—Ä–æ–ø—É—Å–∫–∏ –∏–ª–∏ —É–¥–∞–ª—è—Ç—å —Å—Ç—Ä–æ–∫–∏ —Å –Ω–∏–º–∏ –∏–∑ df - –≤—Å–µ–≥–¥–∞ –∏—â–∏ —ç–∫—Å–ø–µ—Ä—Ç–∞ –≤ –ø—Ä–µ–¥–º–µ—Ç–Ω–æ–π –æ–±–ª–∞—Å—Ç–∏ –∏ —Å–æ–≤–µ—Ç—É–π—Å—è —Å –Ω–∏–º. –ö–∞–∫ –ø—Ä–∞–≤–∏–ª–æ —ç—Ç–æ —á–µ–ª–æ–≤–µ–∫ –∏–∑ –±–∏–∑–Ω–µ—Å–∞, –∫–æ—Ç–æ—Ä—ã–π –∑–Ω–∞—á–∏—Ç —Å—É—Ç—å –±–∏–∑–Ω–µ—Å-–ø—Ä–æ—Ü–µ—Å—Å–∞ –∏ –≤–æ—Ç –∏–º–µ–Ω–Ω–æ –æ–Ω –¥–∞—Å—Ç —Ç–µ–±–µ —Ç–æ—á–Ω—ã–π –∏ –∞—Ä–≥—É–º–µ–Ω—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π —Å–æ–≤–µ—Ç - —á—Ç–æ –≤ –¥–∞–Ω–Ω–æ–π, –∫–æ–Ω–∫—Ä–µ—Ç–Ω–æ–π –∑–∞–¥–∞—á–µ –¥–µ–ª–∞—Ç—å —Å "–±–∏—Ç—ã–º–∏" –ø–æ–ª—è–º–∏ –≤ —Å—Ç—Ä–æ–∫–∞—Ö –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞.</div>


–ü–æ—Å–º–æ—Ç—Ä–∏–º, —Å–∫–æ–ª—å–∫–æ –≤—Å–µ–≥–æ –ø—Ä–æ–ø—É—Å–∫–æ–≤ –≥–æ–¥–∞ –≤—ã–ø—É—Å–∫–∞.

```python
'–í—Å–µ–≥–æ –ø—Ä–æ–ø—É—Å–∫–æ–≤ –≥–æ–¥–∞ –≤—ã–ø—É—Å–∫–∞: {:.1%}'.format((df['year_of_release'].isna().sum())/len(df))
```

–í–æ—Å—Å—Ç–∞–Ω–æ–≤–∏–º —á–∞—Å—Ç—å –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –≥–æ–¥–æ–≤ –ø–æ —Å–ª–µ–¥—É—é—â–∏–º –ø—Ä–∞–≤–∏–ª–∞–º:
- –µ—Å–ª–∏ —É –∏–≥—Ä—ã –µ—Å—Ç—å –≤–µ—Ä—Å–∏—è –¥–ª—è –¥—Ä—É–≥–∏—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º, –≤–æ–∑—å–º—ë–º –≥–æ–¥ –≤—ã–ø—É—Å–∫–∞ –æ—Ç—Ç—É–¥–∞;
- –∏–Ω–∞—á–µ, –ø–æ–ø—Ä–æ–±—É–µ–º –≤–æ—Å—Å—Ç–∞–Ω–æ–≤–∏—Ç—å –≥–æ–¥ –≤—ã–ø—É—Å–∫–∞ –∏–∑ –ø–æ–ª—è `name`, —Ç.–∫. –∏–Ω–æ–≥–¥–∞ –æ–Ω —É–∫–∞–∑—ã–≤–∞–µ—Ç—Å—è –≤ —Å–∞–º–æ–º –Ω–∞–∑–≤–∞–Ω–∏–∏ –∏–≥—Ä—ã.

–ù–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è –±—É–¥–µ—Ç –≤–æ—Å—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞—Ç—å –≥–æ–¥–∞, –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ –ø—Ä–∏–º–µ–Ω—è—è –ø—Ä–∞–≤–∏–ª–∞ –≤—ã—à–µ.

```python
def find_year(row):
    year_of_release = row['year_of_release']
    if pd.isna(year_of_release):
        try:
            year_of_release = df[(df['name']==row['name'])&
                              ~(df['year_of_release'].isna())]['year_of_release'].max()
        except:
            if re.search('20..|19..', row['name']):
                year_of_release = int(re.findall('20..|19..', row['name'])[0])
    return year_of_release
        
```

```python
df['year_of_release']=df.apply(find_year, axis=1)
```

"–ü–æ–≥–æ–¥—ã" –æ–Ω–∏ –Ω–∞–º –Ω–µ —Å–¥–µ–ª–∞—é—Ç, –∞ –Ω–∞–ø–∏—Å–∞–Ω–∏–µ —Ç–∞–∫–æ–≥–æ –∫–æ–¥–∞ –≤—Å—ë —Ä–∞–≤–Ω–æ —Ç—Ä–µ–±—É–µ—Ç –≤—Ä–µ–º–µ–Ω–∏ –Ω–∞ –Ω–∞–ø–∏—Å–∞–Ω–∏–µ –∏ –æ—Ç–ª–∞–¥–∫—É.

–ö–∞—Ç–µ–≥–æ—Ä–∏–∑–∏—Ä—É–µ–º –∑–Ω–∞—á–µ–Ω–∏—è —Ä–µ–π—Ç–∏–Ω–≥–∞ –≤ –∫–æ–ª–æ–Ω–∫–µ `rating`.

–°–æ–≥–ª–∞—Å–Ω–æ –í–∏–∫–∏–ø–µ–¥–∏–∏, —Å—É—â–µ—Å—Ç–≤—É—é—Ç —Å–ª–µ–¥—É—é—â–∏–µ –∑–Ω–∞–∫–∏ —Ä–µ–π—Ç–∏–Ω–≥–∞ ESRB:
- `EC` (Early childhood) ‚Äî –∏–≥—Ä–∞ –ø–æ–¥—Ö–æ–¥–∏—Ç –¥–ª—è –¥–µ—Ç–µ–π –æ—Ç 3 –ª–µ—Ç
- `E` (Everyone) ‚Äî –¥–ª—è –≤—Å–µ—Ö. –ü–µ—Ä–≤–æ–Ω–∞—á–∞–ª—å–Ω–æ `K-A` (Kids to Adults)
- `E10+` (Everyone 10 and older) ‚Äî –¥–ª—è –≤—Å–µ—Ö –æ—Ç 10 –ª–µ—Ç –∏ —Å—Ç–∞—Ä—à–µ
- `T` (Teen) ‚Äî –¥–ª—è –ª–∏—Ü –æ—Ç 13 –ª–µ—Ç
- `M` (Mature) ‚Äî –¥–ª—è –≤–∑—Ä–æ—Å–ª—ã—Ö, –º–∞—Ç–µ—Ä–∏–∞–ª—ã –∏–≥—Ä—ã –Ω–µ –ø–æ–¥—Ö–æ–¥—è—Ç –¥–ª—è –ø–æ–¥—Ä–æ—Å—Ç–∫–æ–≤ –º–ª–∞–¥—à–µ 17 –ª–µ—Ç
- `AO` (Adults Only 18+) ‚Äî —Ç–æ–ª—å–∫–æ –¥–ª—è –≤–∑—Ä–æ—Å–ª—ã—Ö —Å—Ç–∞—Ä—à–µ 18 –ª–µ—Ç
- `RP` (Rating Pending) ‚Äî —Ä–µ–π—Ç–∏–Ω–≥ –æ–∂–∏–¥–∞–µ—Ç—Å—è: –ø—Ä–æ–¥—É–∫—Ç –±—ã–ª –æ—Ç–ø—Ä–∞–≤–ª–µ–Ω –≤ ESRB –∏ –æ–∂–∏–¥–∞–µ—Ç –ø—Ä–∏—Å–≤–æ–µ–Ω–∏—è —Ä–µ–π—Ç–∏–Ω–≥–∞. 


1. `EC`, `E`, `K-A` –∏ `E10+` –æ–±—ä–µ–¥–∏–Ω–∏–º –≤ –æ–¥–Ω—É –∫–∞—Ç–µ–≥–æ—Ä–∏—é `E`
2. `AO` –≤–∫–ª—é—á–∏–º –≤ `M`
3. `RP` –∑–∞–º–µ–Ω–∏–º –Ω–∞ –∑–Ω–∞—á–µ–Ω–∏–µ `unknown`.

```python
df['rating'] = df['rating'].replace('AO','M')
df['rating'] = df['rating'].replace('E10+','E')
df['rating'] = df['rating'].replace('EC','E')
df['rating'] = df['rating'].replace('K-A','E')
df['rating'] = df['rating'].replace('RP','unknown')
```

–ù–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è —á–∞—Å—Ç–∏—á–Ω–æ –∑–∞–ø–æ–ª–Ω–∏—Ç –æ—Ç—Å—É—Ç—Å—Ç–≤—É—é—â–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ –ø–æ —Å–ª–µ–¥—É—é—â–∏–º –ø—Ä–∞–≤–∏–ª–∞–º:
- –µ—Å–ª–∏ —É –∏–≥—Ä—ã –µ—Å—Ç—å –≤–µ—Ä—Å–∏—è –¥–ª—è –¥—Ä—É–≥–∏—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º, –≤–æ–∑—å–º—ë–º —Ä–µ–π—Ç–∏–Ω–≥ –æ—Ç—Ç—É–¥–∞;
- –æ—Å—Ç–∞–ª—å–Ω—ã–µ –ø—Ä–æ–ø—É—Å–∫–∏ –∑–∞–º–µ–Ω–∏–º –Ω–∞ `unknown`.

```python
def find_rating(row):
    rating = row['rating']
    if pd.isna(rating):
        try:
            rating = df[(df['name']==row['name'])&
                              ~(df['rating'].isna())]['rating'].iloc[0]
        except:
            rating = 'unknown'
    return rating
```

```python
df['rating'] = df.apply(find_rating, axis=1)
```

### –ü–æ–ø—É–ª—è—Ä–Ω—ã–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã

```python
df_2015.pivot_table(
    index='year_of_release',
    columns='platform',
    values='total_sales',
    aggfunc='sum').plot(
        title='–ü—Ä–æ–¥–∞–∂–∏ –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º',
        style='o-',
        figsize=(16, 8),
        alpha=0.6,
        linewidth=3,
        xlabel='–ì–æ–¥ –≤—ã–ø—É—Å–∫–∞',
        ylabel='–ü—Ä–æ–¥–∞–∂–∏, –º–ª–Ω —à—Ç')
plt.show()
```
